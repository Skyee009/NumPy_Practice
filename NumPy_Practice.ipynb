# 1. Array Creations

```python
import numpy as np
```

```python
# NumPy stands for Numerical Python 
# Used for...
# Fast numerical computation, working with arrays and matrices, scientific computing, ML Data Science
```

# Easy

```python
# Creating a NumPy array
arr = np.array([1,2,3,4,5])
```

```python
# Cretaing a NumPy array with zeroes, lenght 5
arr1 = np.zeros(5)
print(arr1)
```

```python
# Create a NumPy array with ones, length 4
arr2 = np.ones(4)
arr2
```

# Medium

```python
# create a numpy array starting from 10 to 50 with a step of 10
arr3 = np.arange(10, 51, 10)
arr3
```

```python
# Create a NumPy array of 5, equally spaced values between 0 and 1
arr4 = np.linspace(0, 1, 5)
arr4
# Creates equal gaps 
```

```python
# Create a (3 x 3) NumPy array filled with value 9
arr5 = np.full((3, 3), 9)
arr5
```

# Hard

```python
# Create NumPy array (4 x 4), with integers inbetween 1 to 100
arr6 = np.random.randint(1, 100, size = (4, 4))
arr6
```

```python
# Create a (5 x 5) identity matrix
arr7 = np.eye(5)
arr7
```

```python
# create a 3D NumPy array of shape (2, 3, 4) with random floating-point numbers
arr8 = np.random.rand(2, 3, 4)
arr8
```
